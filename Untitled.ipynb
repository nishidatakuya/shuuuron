```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn import preprocessing, model_selection
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import os
import glob
import re

class Matrix:
    global path
    path = glob.glob("/Users/*/Desktop/*/データ")[0]
    def  __init__(self, date, direction, use_r, target_r, use_value, interval, RW):
        
        self.date = date
        self.direction = direction
        self.use_r = use_r
        self.target_r = target_r
        self.use_value = use_value
        self.interval = interval
        self.RW = RW
        
    def data_din(self):
        
        os.chdir(path+"/OutputsRDIP_TEPS")
        
        #FDAのデータ読み込み
        df_FDA = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTFDA_RDIP"+self.direction+".csv", 
                             names = ("Call sign", "Dep.airport", "Arr.airport/RW", "Aircraft type", "Gate name", "ATA@gate", "Arr.airport", "ATA@airport"),
                             usecols = ["Call sign", "Dep.airport", "Aircraft type", "Gate name", "ATA@airport"])
        df_FDA.insert(0,"Date", self.date)
        #RECATに分類してFDAに追加
        df_FDA = df_FDA.set_index("Call sign", drop = True).rename_axis("Call sign")
        df_RECAT = pd.read_csv(path+"/OutputsFDA/recat.csv").set_index("ICAO TYPE", drop = True)
        df_FDA["RECAT"] = "0"
        
        for callsign in df_FDA.index.values:
            if df_FDA.at[callsign, "Aircraft type"] in df_RECAT.index.values:
                df_FDA.at[callsign, "RECAT"] = df_RECAT.at[df_FDA.at[callsign, "Aircraft type"], "RECAT-EU"]
                
        return df_FDA
    
    def data_cin(self):

        #各種量的変数の取り込み(TEPS)
        os.chdir(path+"/OutputsRDIP_TEPS")
            
        df_FTD = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTFTD_RDIP"+self.direction+".csv").set_index("Call sign", drop = True).rename_axis("Call sign")
        df_GSD = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTGSD_RDIP"+self.direction+".csv").set_index("Call sign", drop = True).rename_axis("Call sign")
        df_ALT = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTALTD_RDIP"+self.direction+".csv").set_index("Call sign", drop = True).rename_axis("Call sign")
        df_TSS = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTTSS2_RDIP"+self.direction+".csv").set_index("Call sign", drop = True).rename_axis("Call sign")
        df_ANG = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTANG2RJTT_RDIP"+self.direction+".csv").set_index("Call sign", drop = True).rename_axis("Call sign")
        df_TSD = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTTSD_RDIP"+self.direction+".csv").set_index("Call sign", drop = True).rename_axis("Call sign")
        
        if self.use_r[0] <= 80 or min(self.target_r) <= 80:
            os.chdir(path+"/OutputsRDIP_TAPS")
            
            #print(glob.glob(self.date + "/" + self.date + "GSD" + "*.csv"))
            df_GSD_TAPS = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTGSD_RDIP_RJTT"+str(self.RW)+".csv").set_index("Call sign", drop = True).rename_axis("Call sign")
            df_ALT_TAPS = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTALTD_RDIP_RJTT"+str(self.RW)+".csv").set_index("Call sign", drop = True).rename_axis("Call sign")
            df_TSS_TAPS = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTTSS2_RDIP_RJTT"+str(self.RW)+".csv").set_index("Call sign", drop = True).rename_axis("Call sign")
            df_ANG_TAPS = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTANG2_RDIP_RJTT"+str(self.RW)+".csv").set_index("callsign", drop = True).rename_axis("Call sign")
            df_TSD_TAPS = pd.read_csv(self.date+"xxxxRJTT/"+self.date+"xxxxRJTTTSD_RDIP_RJTT"+str(self.RW)+".csv").set_index("Call sign", drop = True).rename_axis("Call sign")
            #print(df_GSD_TAPS)
            df_FTD_TAPS = pd.DataFrame(columns = df_GSD_TAPS.columns.values, index = df_GSD_TAPS.index.values)
            for callsign in df_FTD_TAPS.index.values:
                for clm in np.delete(df_FTD_TAPS.columns.values, [-1, -2]):
                    df_FTD_TAPS.at[callsign, clm] = df_TSD_TAPS.at[callsign, "5"] - df_TSD_TAPS.at[callsign, clm]
            df_FTD_TAPS = df_FTD_TAPS.rename_axis("Call sign")
        #print(df_FTD_TAPS)
        df_c = [df_FTD, df_GSD, df_ALT, df_TSS, df_ANG, df_TSD]
        df_c_taps = [df_FTD_TAPS, df_GSD_TAPS, df_ALT_TAPS, df_TSS_TAPS, df_ANG_TAPS, df_TSD_TAPS]
        
        if self.use_r[0] < 80 or min(self.target_r) < 80:
            for i in range(len(df_c_taps)):
                for callsign in df_c[i].index.values:
                    if callsign in df_c_taps[i].index.values:
                        for clm in [clms_c1 for clms_c1 in df_c[i].columns.values[0:-1] if int(re.search(r"\d+", clms_c1).group()) < 80]:
                            df_c[i].at[callsign, clm] = df_c_taps[i].at[callsign, clm]
                    else:
                        df_c[i] = df_c[i].drop(callsign)
        return df_c
    
    def matrix_d(self):
        
        df_FDA = self.data_din()
        clms = []
        for value in self.use_value.values():
            clms = clms + value
        df_matrix_d = pd.DataFrame(columns = clms, index = df_FDA.index.values)
        
        for callsign in df_matrix_d.index.values:
            #該当するエアラインの値を１にする
            if callsign[0:3] in self.use_value["Airline"]:
                df_matrix_d.at[callsign, callsign[0:3]] = 1
            else:
                df_matrix_d.at[callsign,"other_airline"] = 1
            #該当する通過ゲート名を１にする
            if df_FDA.at[callsign,"Gate name"] in self.use_value["Gate"]:
                df_matrix_d.at[callsign,df_FDA.at[callsign,"Gate name"]] = 1
            #else:
                #if self.direction == "N":
                #df_matrix_d.at[callsign, "other_gate"] = 1
            #該当する出発空港(エリア)を１にする
            
            if df_FDA.at[callsign,"Dep.airport"][0] in ["E","L", "U"]:
                df_matrix_d.at[callsign,"EU"] = 1
            elif df_FDA.at[callsign,"Dep.airport"][0] in ["P","A"]:
                if df_FDA.at[callsign,"Dep.airport"][0:2] == "PA":
                    df_matrix_d.at[callsign, "AMERICA"] = 1
                else:
                    df_matrix_d.at[callsign,"PACIFIC"] = 1
            elif df_FDA.at[callsign,"Dep.airport"][0] in ["K","C"]:
                df_matrix_d.at[callsign,"AMERICA"] = 1
            elif df_FDA.at[callsign,"Dep.airport"][0] in ["Z","V","O","W"]:
                if df_FDA.at[callsign,"Dep.airport"][0:2] == "ZM":
                    df_matrix_d.at[callsign, "EU"] = 1
                else:
                    #if self.direction == "S":
                    df_matrix_d.at[callsign,"ASIA"] = 1
            elif df_FDA.at[callsign,"Dep.airport"] in ["YSSY","NZAA"]:
                df_matrix_d.at[callsign,"OCEANIA"] = 1
            #elif df_FDA.at[callsign,"Dep.airport"][0] in ["U"]:
                #df_matrix_d.at[callsign,"RUSSIA"] = 1
            if df_FDA.at[callsign,"Dep.airport"][0] == "R":
                if df_FDA.at[callsign,"Dep.airport"][0:2] in ["RJ", "RO"]:
                    if df_FDA.at[callsign,"Dep.airport"] in self.use_value["Airport_Japan"]:
                        df_matrix_d.at[callsign,df_FDA.at[callsign,"Dep.airport"]] = 1
                    elif df_FDA.at[callsign,"Dep.airport"] in ["RJOO", "RJBB"]:
                        if "RJOOBB" in df_matrix_d.columns.values:
                            df_matrix_d.at[callsign,"RJOOBB"] = 1
                        else:
                            df_matrix_d.at[callsign, "other_JAPAN"] = 1
                    #elif df_FDA.at[callsign,"Dep.airport"][2] in ["S", "C", "E"]:
                        #df_matrix_d.at[callsign, "other_JAPAN_N"] = 1
                    else:
                        df_matrix_d.at[callsign,"other_JAPAN"] = 1
                else:
                    #if self.direction == "S":
                    df_matrix_d.at[callsign,"ASIA"] = 1
            #該当する機体のRECAT区分を１にする
            if df_FDA.at[callsign,"RECAT"] in self.use_value["Recat"]:
                df_matrix_d.at[callsign,df_FDA.at[callsign,"RECAT"]] = 1
            #該当する空港到着時間帯を１にする
            if re.sub("[0-9]", "", str(df_FDA.at[callsign, "ATA@airport"])) == "":
                a = f'{int(df_FDA.at[callsign, "ATA@airport"])+900:04}'
                if int(a) > 2400:
                    a = f'{int(a)-2400:04}'
                a2 = int(df_FDA.at[callsign, "ATA@airport"]) % 100
                
                for clm in self.use_value["Ata_JST"]:
                    b = re.search(r"\d+(\.\d+)?", clm).group()
                    c = clm.replace(b+"-", "")
                    if int(a[0:2])+(a2/60) >= float(b) and int(a[0:2])+(a2/60) < float(c):
                        df_matrix_d.at[callsign,clm] = 1
        #１にならなかった部分を０とし、１が５つない行を削除
        df_matrix_d = df_matrix_d.fillna(0)
        #df_matrix_d = df_matrix_d[df_matrix_d.sum(axis = 0) > 1]
        df_matrix_d = df_matrix_d[df_matrix_d.sum(axis = 1) == len(self.use_value) - 1]
        return df_matrix_d
        
    def matrix_c(self):
        
        
        df_c = self.data_cin()
        df_FDA = self.matrix_d()
    
        a = self.use_r[0]  #200
        b = self.use_r[1]  #400
        c = self.use_r[2]  #20
        d = self.target_r[0]  #80 
        e = self.target_r[1]  #200
        clms = []
        section = np.arange(d, a+0.5, self.interval)
    
        #各量的変数の列名を作成
        for i in range(b-c,a-c,-c):
            clms.append(str(i+c)+"TO"+str(i)+"FT")
        for i in range(b, a-c, -c):
            clms.append(str(i)+"GS")
        for i in range(b, a-c, -c):
            clms.append(str(i)+"ALT")
        for i in range(b, a-c, -c):
            clms.append(str(i)+"SEP")
        for i in range(b, a-c, -c):
            clms.append(str(i)+"ANG")
            
        clms.append("Atm")
        clms.append("Pre")
        clms.append("Temp")
        
        for i in range(len(section)-1):
            clms.append(str(int(section[i+1]))+"TO"+str(int(section[i]))+"NUM")
        #if section[-1] < a:
            #clms.append(str(int(section[-1]))+"TO"+str(a)+"NUM")
        clms.append(str(a) + "NM" + "_5MIN_AGO")
        clms.append(str(a) + "NM" + "_10MIN_AGO")
        clms.append(str(a) + "NM" + "_15MIN_AGO")
        clms.append(str(a) + "NM" + "_5MIN_LATER")
        clms.append(str(a) + "NM" + "_10MIN_LATER")
        clms.append(str(a) + "NM" + "_15MIN_LATER")
        clms.append("PREC")
        
        clms.append(str(e) + "TO" + str(d) + "FT")
        clms.append(str(e) + "TSD")
        clms.append(str(d) + "TSD")
        
        
        index0 = []
        for index1 in df_FDA.index.values:
            if index1 in df_c[0].index.values:
                index0.append(index1)
        
        #print(clms)
        df_matrix_c = pd.DataFrame(columns = clms, index = index0)
        #print(df_matrix_c)
        #各種量的変数の値を代入(外れ値と思われるものは代入しない)
        df_crowd = pd.DataFrame(index = index0)
        for callsign in df_matrix_c.index.values:
            
            for clm in [strng for strng in clms if strng[-2:] == "FT"]:
                A = re.search(r"\d+", clm).group()
                B = clm.replace(re.search(r"\d+", clm).group()+"TO", "").replace("FT","") #"A TO B FT"
                
                num = df_c[0].at[callsign,A] - df_c[0].at[callsign,B]
                
                if num > 0:
                    
                    if a >= 150:
                        if (int(A) - int(B)) / 20 * 800 > num:
                            df_matrix_c.at[callsign, clm] = df_c[0].at[callsign,A] - df_c[0].at[callsign,B]
                    elif a < 150:
                        #if (int(A) - int(B)) / 20  * 800 > num:
                    
                        df_matrix_c.at[callsign, clm] = df_c[0].at[callsign,A] - df_c[0].at[callsign,B]
                        
            for clm in [strng for strng in clms if strng[-2:] == "GS"]:

                num = re.search(r"\d+", clm).group()
                
                if df_c[1].at[callsign,re.search(r"\d+", clm).group()] > 0:
                    df_matrix_c.at[callsign, clm] = df_c[1].at[callsign,num]
                        
            for clm in [strng for strng in clms if strng[-3:] == "ALT"]:

                num = re.search(r"\d+", clm).group()
                if df_c[2].at[callsign,re.search(r"\d+", clm).group()] > 0:
                    df_matrix_c.at[callsign, clm] = df_c[2].at[callsign,num]
                        
            for clm in [strng for strng in clms if strng[-3:] == "SEP"]:

                num = re.search(r"\d+", clm).group()
                if df_c[3].at[callsign,num] > 0 and df_c[3].at[callsign,num] < 50000:
                    df_matrix_c.at[callsign, clm] = df_c[3].at[callsign,num]
                        
            for clm in [strng for strng in clms if strng[-3:] == "ANG"]:

                num = re.search(r"\d+", clm).group()
                if df_c[4].at[callsign,num+"_Ang"] > 0 and df_c[4].at[callsign,num+"_Ang"] <= 360:
                    df_matrix_c.at[callsign, clm] = df_c[4].at[callsign, str(num)+"_Ang"]
                    
            df_we = pd.read_csv(path+"/気象/weather.csv", index_col = 0)
            
            for callsign5 in df_matrix_c.index.values:
                t =  df_c[5].at[callsign5, str(self.use_r[0])]
                Date = self.date[0:4] + "/" + str(int(self.date[4:6])) + "/" + str(int(self.date[6:8]))
                if t <= 86400 and t > 0:
                    Hour = int(t // 3600) + 1
                    if Hour % 24 == 0:
                        Hour = "00"
                        Date = self.date[0:4] + "/" + str(int(self.date[4:6])) + "/" + str(int(self.date[6:8])+1)
                    df_matrix_c.at[callsign5, "Atm"] = df_we.at[Date+" "+str(Hour)+":00:00", "Atm"]
                    df_matrix_c.at[callsign5, "Pre"] = df_we.at[Date+" "+str(Hour)+":00:00", "Pre"]
                    df_matrix_c.at[callsign5, "Temp"] = df_we.at[Date+" "+str(Hour)+":00:00", "Temp"]
                
            
            ATA = df_c[5].at[callsign,str(a)]
                #print(callsign,ATA)
                
            for i in range(len(section)-1):
                start = int(section[i])
                end   = int(section[i+1])
                count_bool = ((df_c[5][str(end)] < ATA) & (df_c[5][str(start)] > ATA) & (df_c[5][str(a)] != ATA))
                df_matrix_c.at[callsign, str(end)+"TO"+str(start)+"NUM"] = count_bool.sum()
                df_crowd.at[callsign, str(end)+"TO"+str(start)+"NUM"] = count_bool.sum()
                
            for clm in [strng for strng in clms if re.search("NM_" + r"\d+" + "MIN_AGO", strng) != None]:
                Min = int(re.search(r"\d+", clm[-10:-1]).group())
                count_bool2 = ((df_c[5][str(a)] > ATA - Min*60) & (df_c[5][str(a)] < ATA))
                df_matrix_c.at[callsign, clm] = count_bool2.sum()
                df_crowd.at[callsign, clm] = count_bool2.sum()
                
            for clm in [strng for strng in clms if re.search("NM_" + r"\d+" + "MIN_LATER", strng) != None]:
                Min = int(re.search(r"\d+", clm[-12:-1]).group())
                count_bool3 = ((df_c[5][str(a)] < ATA + Min*60) & (df_c[5][str(a)] > ATA))
                df_matrix_c.at[callsign, clm] = count_bool3.sum()
                df_crowd.at[callsign, clm] = count_bool3.sum()
                    
            
            callsign2 = df_matrix_c.index.values[0]
            ATA2 = df_c[5].at[callsign2,str(a)]
            for Callsign2 in df_c[5].index.values:
                if df_c[5].at[Callsign2,str(a)] > ATA2 and df_c[5].at[Callsign2,str(a)] < ATA:
                    callsign2 = Callsign2
                    ATA2 = df_c[5].at[Callsign2,str(a)]
            df_matrix_c.at[callsign, "PREC"] = df_c[0].at[callsign2, str(a)]
            
            df_matrix_c.at[callsign, str(self.target_r[1])+"TSD"] = df_c[5].at[callsign, str(e)]
            df_matrix_c.at[callsign, str(self.target_r[0])+"TSD"] = df_c[5].at[callsign, str(d)]
                
            #else:
                #print(callsign, "ERROR!!!")
                
        #print(df_matrix_c)

        #値を取得できなかったものを含む行を削除
        df_matrix_c = df_matrix_c.dropna(how = "any", axis = 0).dropna(how = "all", axis = 1)
        """
        if self.date == "20190911":
            os.chdir(path)
            os.chdir("../")
            path2 = os.getcwd()
            df_matrix_c.to_csv(path2+"/結果/test_ver2.csv")
        """    
        
        df_matrix_c.to_csv(path+"/混雑/"+self.date+self.RW+".csv")
        #print(df_matrix_c.index.values)
        #値がおかしいAPJ808, SIA630, APJ860, JAL922を削除
        for callsign in df_matrix_c.index.values:
            if callsign in ["APJ808 ", "SIA630 ", "APJ860 ", "JAL922 ", "VJC936 ", "UAE312 "]:
                df_matrix_c = df_matrix_c.drop(callsign)
                #print(callsign,"削除ずみ")
        #df_matrix_c = df_matrix_c[df_matrix_c[str(e)+"TO"+str(d)+"FT"] < 1500]
        #値を標準化  
        #print(df_matrix_c)
        os.chdir(path+"/OutputsRDIP_TEPS")
        X_c = np.delete(df_matrix_c.values, [-1, -2, -3], 1)
        ss = preprocessing.StandardScaler()
        ss.fit(X_c)
        X_css = ss.transform(X_c)
        clms2 = np.delete(df_matrix_c.columns.values, [-1, -2, -3])
        df_matrix_c[clms2] = pd.DataFrame(X_css,columns = clms2, index = df_matrix_c.index.values)
        return df_matrix_c
    """
        def matrix_weather(self):
            we = Matrix(re.sub("[^0-9]", "", self.date), self.direction, self.use_r, self.target_r, self.use_value, self.interval)
            df_we = pd.read_csv("/Users/nishidatakuya/Desktop/卒論/気象データ/weather.csv", index_col = 0)

        for callsign in df_matrix_weather.index.values:
            t =  df_c.at[callsign, str(self.use_r[0])]
            Date = self.date[0:4] + "/" + str(int(self.date[4:6])) + "/" + str(int(self.date[6:8]))
            if t <= 86400 and t > 0:
                Hour = int(t // 3600) + 1
                if Hour % 24 == 0:
                    Hour = "00"
                    Date = self.date[0:4] + "/" + str(int(self.date[4:6])) + "/" + str(int(self.date[6:8])+1)
                df_matrix_weather.at[callsign, "Atm"] = df_we.at[Date+" "+str(Hour)+":00:00", "Atm"]
                df_matrix_weather.at[callsign, "Pre"] = df_we.at[Date+" "+str(Hour)+":00:00", "Pre"]
                df_matrix_weather.at[callsign, "Temp"] = df_we.at[Date+" "+str(Hour)+":00:00", "Temp"]
            
        df_matrix_weather = df_matrix_weather.dropna(how = "any", axis = 0)
        
        ss = preprocessing.StandardScaler()
        ss.fit(df_matrix_weather)
        df_matrix_weather_ss = pd.DataFrame(ss.transform(df_matrix_weather), 
                                            columns = df_matrix_weather.columns.values, 
                                            index = df_matrix_weather.index.values)
        
        return df_matrix_weather_ss
"""    
    
    
    def matrix_all(self):
    
        df_d = self.matrix_d()
        df_c = self.matrix_c()
        
        #print("df_c", df_c)
            
        df_all = pd.merge(df_d, df_c, left_index = True, right_index = True, how = "inner")
        
        return df_all

class Integ(Matrix):
    def __init__(self, date, direction, use_r, target_r, use_value, interval, RW):
        super().__init__(date, direction, use_r, target_r, use_value, interval, RW)
        
    def integrate(self):
        
        if self.use_r[0] >= 80 and min(self.target_r) >= 80:
            os.chdir(path+"/OutputsRDIP_TEPS")
        else:
            os.chdir(path+"/OutputsRDIP_TAPS")
        folderlist = sorted(glob.glob("*"))
        folderlist2 = sorted(glob.glob("*/*"+self.RW+"*.csv"))
        df_int = pd.DataFrame()
        for folder in folderlist:
            self.date = re.sub("[^0-9]", "", folder)
            if folder+"/"+folder+"FDA_RDIP_RJTT"+self.RW+".csv" in folderlist2:
                if (self.date not in ["20191113", "20191212", "20200205", "20200207"]) or self.RW != "23":
                    df = super().matrix_all()
                    df_int = pd.concat([df_int,df])
            else:
                print("No Data in "+ self.date)
            
        return df_int
    
```
